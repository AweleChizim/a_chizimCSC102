```python
#project 3
#import pandas
import pandas as pd

name = input("Name:")
age = input("Age:")
department = int(input("Enter 1 for Computer Science or 2 for Mass Communication:"))


def csc_check():
    jamb_score = int(input("JAMB Score:"))
    print("KEY:\nEnter 1 for A1 \n      2 for B2&3\n      3 for C4-6\n      4 for D7\n      5 for E8\n      6 for F9")
    math = int(input("MATH:"))
    eng = int(input("ENGLISH:"))
    phy = int(input("PHYSICS:"))
    chem = int(input("CHEMISTRY:"))
    fmath = int(input("FURTHER MATH:"))
    interview = int(input("INTERVIEW - Enter 1 for PASS or 2 for FAIL:"))

    if (jamb_score >= 230) and (interview == 1) and (math >= 1 and math <= 3) and (eng >= 1 and eng <= 3) and (phy >= 1 and phy <= 3) and (chem >= 1 and chem <= 3) and (fmath >= 1 and fmath <= 3):
        status = "ADMITTED"
        data = {'Name': [name], 'Department': ["Computer Science"], 'Age': [age]}
        ad = pd.DataFrame(data)
        ad.to_csv('admitted.csv')
    else:
        status = "NOT ADMITTED"
        data = {'Name': [name], 'Department': ["Computer Science"], 'Age': [age]}
        ad = pd.DataFrame(data)
        ad.to_csv('not-admitted.csv')
    return print(f'{status}')

def masscom_check():
    jamb_score = int(input("JAMB Score:"))
    print("KEY:\nEnter 1 for A1 \n      2 for B2&3\n      3 for C4-6\n      4 for D7\n      5 for E8\n      6 for F9")
    math = int(input("MATH:"))
    eng = int(input("ENGLISH:"))
    crs = int(input("CRS:"))
    lit = int(input("LITERATURE:"))
    gov = int(input("GOVERNMENT:"))
    interview = int(input("INTERVIEW - Enter 1 for PASS or 2 for FAIL:"))

    if (jamb_score >= 220) and (interview == 1) and (math >= 1 and math <= 3) and (eng >= 1 and eng <= 3) and (crs >= 1 and crs <= 3) and (lit >= 1 and lit <= 3) and (gov >= 1 and gov <= 3):
        status = "ADMITTED"
        data = {'Name': [name], 'Department': ["Mass Communication"], 'Age': [age]}
        ad = pd.DataFrame(data)
        ad.to_csv('admitted.csv')
    else:
        status = "NOT ADMITTED"
        data = {'Name': [name], 'Department': ["Mass Communication"], 'Age': [age]}
        ad = pd.DataFrame(data)
        ad.to_csv('not-admitted.csv')
    return print(f'{status}')

if department == 1:
    csc_check()
elif department == 2:
    masscom_check()
```
